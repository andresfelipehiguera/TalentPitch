# Video Stats

- https://www.edlitera.com/blog/posts/emotion-detection-in-images

```python
import matplotlib.pyplot as plt
import imageio.v3 as iio
from deepface import DeepFace
```

```python
frame = iio.imread(
    filename,
    index=100,
    plugin="pyav",
)

plt.imshow(frame)
plt.savefig(filename.split('.')[0] + '.png')
plt.show()
```

```python
face_analysis = DeepFace.analyze(img_path=filename.split('.')[0] + '.png')
pprint(face_analysis[0])
```
